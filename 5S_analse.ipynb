```python
#Ler as tabelas

p1 = pd.read_excel("/content/Auditoria 5S - Almoxarifado 1.xlsx")
p2 = pd.read_excel("/content/Auditoria 5S - Banheiros 1.xlsx")
p3 = pd.read_excel("/content/Auditoria 5S - Copa 1.xlsx")
p4 = pd.read_excel("/content/Auditoria 5S - Dep. de Solventes 1.xlsx")
p5 = pd.read_excel("/content/Auditoria 5S - Depósito de Matérias-Primas 1.xlsx")
p6 = pd.read_excel("/content/Auditoria 5S - Escritório 1.xlsx")
p7 = pd.read_excel("/content/Auditoria 5S - Laboratório Analítico 1.xlsx")
p8 = pd.read_excel("/content/Auditoria 5S - Laboratório P&D Agro 1.xlsx")
p9 = pd.read_excel("/content/Auditoria 5S - Sala Soja 1.xlsx")
```

```python
#Juntar todas e tirar as colunas não utilizadas

df = pd.concat([p1,p2,p3,p4,p5,p6,p7,p8,p9], axis = 1)
df = df.drop(['ID','Hora de início', 'Hora de conclusão', 'Email', 'Nome'], axis = 1)
df = df.drop(columns=[col for col in df.columns if 'N.A.' in df[col].values])
df = df.drop(columns=df.select_dtypes(include='float').columns)
df2 = df.filter(like='Plano de Ação')
df = df.drop(columns=[col for col in df.columns if 'Plano de Ação' in col])
```

```python
# Planos de Ação - Colocar todos os planos de ação em uma tabela, separada pelos setores.
Plan_almox = df2.filter(like='_1')
Plan_almox = Plan_almox.transpose()
Plan_almox['Setor'] = 'Almoxarifado'

Plan_Ban = df2.filter(like='_2')
Plan_Ban = Plan_Ban.transpose()
Plan_Ban['Setor'] = 'Banheiros'

Plan_Copa = df2.filter(like='_3')
Plan_Copa = Plan_Copa.transpose()
Plan_Copa['Setor'] = 'Copa'

Plan_Solv = df2.filter(like='_4')
Plan_Solv = Plan_Solv.transpose()
Plan_Solv['Setor'] = 'Depósito de Solventes'

Plan_MatP = df2.filter(like='_5')
Plan_MatP = Plan_MatP.transpose()
Plan_MatP['Setor'] = 'Depósito de MP'

Plan_Escr = df2.filter(like='_6')
Plan_Escr = Plan_Escr.transpose()
Plan_Escr['Setor'] = 'Escritório'

Plan_LabAna = df2.filter(like='_7')
Plan_LabAna = Plan_LabAna.transpose()
Plan_LabAna['Setor'] = 'Lab Analitico'

Plan_LabPeD = df2.filter(like='_8')
Plan_LabPeD = Plan_LabPeD.transpose()
Plan_LabPeD['Setor'] = 'Lab PeD'

Plan_SalSoj = df2.filter(like='_9')
Plan_SalSoj = Plan_SalSoj.transpose()
Plan_SalSoj['Setor'] = 'Sala Soja PeD'

planos = pd.concat([Plan_almox, Plan_Ban, Plan_Copa, Plan_Solv, Plan_MatP, Plan_Escr, Plan_LabAna, Plan_LabPeD, Plan_SalSoj], axis = 0, ignore_index = True)
```
